```python
a = np.array([3.1, 1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])
```

```python
producto_escalar = np.dot(a, b)
print(producto_escalar)
```

¿Qué se debe realizar para poder multiplicarlos?


Ambos vectores deben tener:

-La misma dimensión

-El mismo número de elementos

```python
producto_punto_a_punto = a * b
print(producto_punto_a_punto)
```

```python
A = np.array([
    [2, -1, -3],
    [4, 1.5, -2.5],
    [7.3, -0.9, 0.2]
])

print("Matriz A:")
print(A)
```

```python
AT = A.T

print("Transpuesta de A:")
print(AT)
```

np.ones()

Crea una matriz llena de unos.

```python
matriz_unos = np.ones((3,3))
print("Matriz de unos:")
print(matriz_unos)
```

np.round()

Redondea valores al número de decimales que indiquemos.

```python
redondeada = np.round(A, 0)  # 0 decimales
print("Matriz redondeada:")
print(redondeada)
```

np.ceil()

Redondea hacia arriba (techo).

```python
ceil_A = np.ceil(A)
print("Matriz con ceil:")
print(ceil_A)
```

np.floor()

Redondea hacia abajo (piso).

```python
floor_A = np.floor(A)
print("Matriz con floor:")
print(floor_A)
```

```python
valor = A[0, 2]
print("Valor primera fila, tercera columna:", valor)
```

```python
segunda_fila = A[1, :]
print("Segunda fila:")
print(segunda_fila)
```
