## j)

```python
# Intervalo
n = np.arange(0, 101)  # n desde 0 hasta 100

# Señal
y = np.sin(np.pi * 0.12 * n)

# Configuración de figura
plt.figure(figsize=(10, 5))

# Gráfica tipo discreta
plt.stem(n, y)

# Detalles estéticos
plt.title("y[n] = sin(pi * 0.12 * n)", fontsize=14)
plt.xlabel("n (muestras)", fontsize=12)
plt.ylabel("Amplitud", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

## k)

```python
# Intervalo discreto
n = np.arange(0, 101)  # n desde 0 hasta 100

# Señal
y2 = np.cos(2 * np.pi * 0.03 * n)

# Gráfica discreta correcta
plt.figure(figsize=(10, 5))
plt.stem(n, y2)

plt.title("y2[n] = cos(2π * 0.03 * n)", fontsize=14)
plt.xlabel("n (muestras)", fontsize=12)
plt.ylabel("Amplitud", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

## l)

```python
# 🔹 Tercera señal: suma
s = y + y2

plt.figure(figsize=(10,5))
plt.stem(n, s)

plt.title("s[n] = y[n] + y2[n]", fontsize=14)
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# 🔹 Cuarta señal: producto punto a punto
t = y * y2

plt.figure(figsize=(10,5))
plt.stem(n, t)

plt.title("t[n] = y[n] · y2[n]", fontsize=14)
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

## m)

```python
# Figura
plt.figure(figsize=(10,5))

# Gráficas con colores distintos
plt.plot(n, y, 'b', label='y[n] = sin(π·0.12n)')
plt.plot(n, y2, 'r', label='y2[n] = cos(2π·0.03n)')

# Etiquetas y título
plt.xlabel("n (muestras)", fontsize=12)
plt.ylabel("Amplitud", fontsize=12)
plt.title("Comparación de señales y[n] y y2[n]", fontsize=14)

# Leyenda
plt.legend()

# Cuadrícula
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

## n)

```python
# Crear figura
plt.figure(figsize=(10,5))

# Graficar ambas señales con colores diferentes
plt.plot(n, s, color='blue', label='s[n] = y[n] + y2[n]')
plt.plot(n, t, color='red', label='t[n] = y[n] · y2[n]')

# Títulos y etiquetas
plt.xlabel("n (muestras)")
plt.ylabel("Amplitud")
plt.title("Comparación de señales s[n] y t[n]")

# Leyenda
plt.legend()

# Cuadrícula
plt.grid(True)

plt.tight_layout()
plt.show()
```
